```python
import pandas as pd
import requests
from io import StringIO

url = "https://raw.githubusercontent.com/fivethirtyeight/data/master/nba-elo/nbaallelo.csv"

response = requests.get(url)
df = pd.read_csv(StringIO(response.text))
pd.set_option('display.max_columns', 23)

df.tail()
```

```python
df_espn = pd.read_csv('/content/drive/MyDrive/PAN/NBA - ESPN Data - ESPN.csv')
df_espn.head()
```

```python
#Results for Only OKC
df_okc = df[df['team_id'] == 'OKC']
df_okc.tail()
```

```python
#2014-15 OKC Results
# Convert to datetime
df['date_game'] = pd.to_datetime(df['date_game'])

# Filter OKC home games within the date range
df_okc = df[
    (df['team_id'] == 'OKC') &
    (df['date_game'] >= '2014-10-29') &
    (df['date_game'] <= '2015-04-15') &
    (df['game_location'] == 'H')
].copy()

# Format the date for display
df_okc['date_game'] = df_okc['date_game'].dt.strftime('%m/%d/%Y')

# Show results
df_okc.head(10)
```

```python
# Merge on game_id
merged_df = pd.merge(df_okc, df_espn, on='game_id', how='inner')
pd.set_option('display.max_columns', 28)
merged_df.head()
```

```python
#Average Player High Score for Wins and Losses
avg_player_high_score = merged_df.groupby('game_result')['team_hs_pts'].mean().round(2).reset_index()

avg_player_high_score.columns = ['game_result', 'avg_player_high_score']

avg_player_high_score.head()
```
